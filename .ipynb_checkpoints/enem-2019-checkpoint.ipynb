### Empregado(a) Doméstico(a)
Se um Estudante tem empregado(a) doméstico(a) em casa pelo menos uma vez por semana, ele apresenta, em média, 100 pontos a mais no ENEM. Talvez não precisar realizar tarefas domésticas libere mais tempo para estudar.


Existe um salto na média entre estudantes que não têm empregado doméstico, e os que têm. Mas, entre aqueles que têm, é menos relevantes quantas vezes por semana o empregado doméstico está presente. Isso sugere um feature engineering: podemos criar uma variável do tipo boolean indicando se o estudante possui ou não empregado doméstico. Assim, ao invés de 4 classes, teremos apenas 2 (Sim e Não).

```python
#box plot: Em sua residência trabalha empregado(a) doméstico(a)? x NU_NOTA_MT
var = 'Q007'
A = 'Nenhum'
B = '1 ou 2 dias'
C = '3 ou 4 dias'
D = 'pelo menos 5 dias'
plotdata = pd.concat([data[dependentVariable], data[var].map({'A': A, 'B': B, 'C': C, 'D': D})], axis=1)
f, ax = plt.subplots(figsize=(10, 6))
fig = sns.boxplot(x=var, y=dependentVariable, data=plotdata, order=[A, B, C, D], showmeans=True)
fig.axis(ymin=320);

ax.set_xticklabels(ax.get_xticklabels(),rotation=0)
ax.set(ylabel="Nota de Matemática")
ax.set(xlabel="")
ax.set(title="Em sua residência trabalha empregado(a) doméstico(a) quantos dias por semana?")

f.savefig("empregado.png", bbox_inches='tight', dpi=600)
```

# Feature Engineering
Vamos criar features com base nos insights que ganhamos até agora. A intenção é que as features criadas sejam úteis na regressão da nota.

A primeira feature que criaremos será max_escol: a maior escolaridade entre o pai e mãe. Essa parece ser uma variável interessante, pois aquilo que faz um maior nível de escolaridade dos pais influenciar a nota da prova do ENEM do filho pode estar presente mesmo que somente dos pais tenha uma escolaridade mais avançada. Além disso, resolveremos o problema de missing data para os estudantes que responderam que não sabem a escolaridade de um dos pais, mas sabem do outro.

A segunda feature que criaremos é a renda per capita, que calcularemos pela razão entre a renda mensal da família e o número de pessoas que residem no domicílio do aluno. Esperamos que essa variável seja interessante, principalmente nos intervalos de renda mais baixos, em que um residente adicional pode consumir recursos críticos de modo a não sobrar para se gastar naquilo que poderia fazer a nota do ENEM aumentar.

Vamos aproveitar para já deixar nossas colunas da forma apropriada para que os algoritmos sejam treinados, pois o LightGBM, por exemplo, aceita somente inteiros, floats e booleans. As variáveis categóricas que possuem um ordenamento natural serão convertidas de letras para números ordenados. Por exemplo, temos 16 classes de renda (de 'A' até 'Q'), sendo que qualquer classe de renda sempre apresenta valor de renda maior que a anterior e menor que a seguinte. Desse modo, converteremos 'A' para 0, 'B' para 1, ..., 'Q' para 16. Já a variável que aponta a raça (branco, negro, etc) não possui ordenamento natural. Portanto vamos convertê-la para variáveis dummie.

```python
X3_labels = X3.copy()

#feature eng 1: max_escol
#primeiro vamos mapear os níveis de escolaridade A, ..., G, H em números. Como H = "não sei", colocaremos como -1
X3_labels['Q001'] = X3_labels['Q001'].map({'A': 0, 'B': 1, 'C': 2, 'D': 3, 'E': 4, 'F': 5, 'G': 6, 'H':-1})
X3_labels['Q002'] = X3_labels['Q002'].map({'A': 0, 'B': 1, 'C': 2, 'D': 3, 'E': 4, 'F': 5, 'G': 6, 'H':-1})
#agora vamos criar uma coluna que é a maior escolaridade entre pai e mae
X3_labels['max_escol'] = X3_labels[['Q001', 'Q002']].max(axis=1)
#e jogar fora as duas colunas originais de escolaridade do pai e da mãe
X3_labels.drop(['Q001', 'Q002'], axis=1, inplace=True)

#X3_labels.head()

#FAZENDO OS LABELS DAS OUTRAS COLUNAS

X3_labels['TP_SEXO'] = X3_labels['TP_SEXO'].map({'M': 0, 'F': 1})
#X3_labels['TP_SEXO'].value_counts()
X3_labels['Q006'] = X3_labels['Q006'].map({'A': 0, 'B': 1, 'C': 2, 'D': 3, 'E': 4, 'F': 5, 'G': 6, 'H':7, 'I': 8, 'J': 9, 'K': 10, 'L': 11, 'M': 12, 'N': 13, 'O': 14, 'P': 15, 'Q': 16})
#X3_labels['Q006'].value_counts()
X3_labels['Q007'] = X3_labels['Q007'].map({'A': 0, 'B': 1, 'C': 2, 'D': 3})
#X3_labels['Q007'].value_counts()
X3_labels['Q010'] = X3_labels['Q010'].map({'A': 0, 'B': 1, 'C': 2, 'D': 3, 'E': 4})
#X3_labels['Q010'].value_counts()
X3_labels['Q021'] = X3_labels['Q021'].map({'A': 0, 'B': 1})
#X3_labels['Q021'].value_counts()

#feature eng 2: renda per capita
rendaDict = {
    0: [0,0],
    1: [0,998],
    2: [998.01, 1497],
    3: [1497.01, 1996],
    4: [1996.01, 2495],
    5: [2495.01, 2994],
    6: [2994.01, 3992],
    7: [3992.01, 4990],
    8: [4990.01, 5988],
    9: [5988.01, 6986],
    10: [6986.01, 7984],
    11: [7984.01, 8982],
    12: [8982.01, 9980],
    13: [9980.01, 11976],
    14: [11976.01, 14970],
    15: [14970.01, 19960],
    16: [19960.01, 9000000]    
}


rendaPerCapitaDict = {
0: [0, 0],
1: [0, 200],
2: [200.01, 400],
3: [400.01, 600],
4: [600.01, 800],
5: [800.01, 1000],
6: [1000.01, 1200],
7: [1200.01, 1400],
8: [1400.01, 1600],
9: [1600.01, 1800],
10: [1800.01, 2000],
11: [2000.01, 2400],
12: [2400.01, 2800],
13: [2800.01, 3200] ,   
14: [3200.01, 3600],
15: [3600.01, 4000],
16: [4000.01, 4800],
17: [4800.01, 5600],
18: [5600.01, 9000000]
}

def calcularRendaPerCapitaMinima(classeRenda, residentes):
    if classeRenda == 0:
        return -1
    else:
        renda = np.mean([rendaDict[classeRenda][0], rendaDict[classeRenda][1]])
        rendaPerCapita = renda/residentes
        return rendaPerCapita

#r = calcularRendaPerCapitaMinima(16, 1)
#print(r)

def calcularclassePerCapita(rendaPerCapita):
    if rendaPerCapita == -1:
        return 0
    else:
        for classe in rendaPerCapitaDict:
            #print(str(classe)+' '+str(rendaPerCapitaDict[classe][0])+' '+str(rendaPerCapitaDict[classe][1]))
            if rendaPerCapita >= rendaPerCapitaDict[classe][0] and rendaPerCapita < rendaPerCapitaDict[classe][1]:
                return classe

def criarClassePerCapita(cols):
    classeRenda = cols[0]
    residentes = cols[1]
    #determina a renda per capita mínima desse estudante
    rendaPerCapitaMinima = calcularRendaPerCapitaMinima(classeRenda, residentes)
    #determina qual é a classe de renda per capita onde aquele aluno deve ser colocado
    if rendaPerCapitaMinima == -1:
        return 0
    else:
        return calcularclassePerCapita(rendaPerCapitaMinima)

#print('rendaPerCapitaMinima', calcularRendaPerCapitaMinima(10,6))
#print('classe per capita', calcularclassePerCapita(calcularRendaPerCapitaMinima(10,6)))
#print('criar classe per capita', criarClassePerCapita([10,6]))

X3_labels['perCapita'] = X3_labels[['Q006', 'Q005']].apply(criarClassePerCapita, axis=1)


#dummies de raça
X3_labels = pd.get_dummies(X3_labels, columns=['TP_COR_RACA'])
#eliminando a coluna de one hot encoding de quem respondeu "não declarado" para raça
X3_labels.drop('TP_COR_RACA_0', axis=1, inplace=True)
#X3_labels.head()
```

Nos gráficos abaixo, vemos a influência das feature que criamos na nota do ENEM. A renda per capita apresenta formato de escada até a 12ª classe de renda, a partir da qual fica relativamente estável. Já a escolaridade máxima entre pai e mãe aponta ganhos suaves na nota à medida que o aluno avança nas primeiras 4 classes. Esse ganho é mais acentuado da 4ª para a 5ª, e da 5ª para a 6ª classes, que correspondem ao caso em que algum dos pais possui escolaridade em nível superior.

```python
#box plot: renda per capita x NU_NOTA_MT
var = 'perCapita'
plotdata = pd.concat([model_data2['NU_NOTA_MT'], X3_labels[var]], axis=1)
f, ax = plt.subplots(figsize=(14, 6))
fig = sns.boxplot(x=var, y='NU_NOTA_MT', data=plotdata)
fig.axis(ymin=320);
ax.set_xticklabels(ax.get_xticklabels(),rotation=80)
ax.set(ylabel="Nota de Matemática")
ax.set(xlabel="")
ax.set(title="Renda per capita")
f.savefig("perCapita.png", bbox_inches='tight', dpi=600)


#box plot: max_escol x NU_NOTA_MT
var = 'max_escol'
A = 'Nunca estudou'
B = 'Não completou 5o ano Fundamental'
C = 'Completou 5o ano Fundamental mas não completou 9o ano'
D = 'Completou 5o ano Fundamental mas não completou Médio'
E = 'Completou Médio mas não completou Faculdade'
F = 'Completou Faculdade mas não completou Pós Graduação'
G = 'Completou Pós Graduação'
H = 'Não sei'
plotdata = pd.concat([model_data2['NU_NOTA_MT'], X3_labels[var].map({0: A, 1: B, 2: C, 3: D, 4: E, 5: F, 6: G, -1: H})], axis=1)
f, ax = plt.subplots(figsize=(14, 6))
fig = sns.boxplot(x=var, y='NU_NOTA_MT', data=plotdata, order=[A, B, C, D, E, F, G])
fig.axis(ymin=320);
ax.set_xticklabels(ax.get_xticklabels(),rotation=80)
ax.set(ylabel="Nota de Matemática")
ax.set(xlabel="")
ax.set(title="Maior escolaridade entre pai e mãe")
f.savefig("max_escol.png", bbox_inches='tight', dpi=400)
```

# Feature Importance
Vamos rodar um XGBoost e um LightGBM para serem nossos modelos de benchmark e para estimarmos a importância relativa entre as features na previsão da Nota do ENEM.

A partir do resultado abaixo, podemos ver que o XGBoost e o LightGBM alcançam métricas de avaliação bastante parecidas: cerca de 92.4 para RMSE e 74.6 para MAE. Esses valores são nosso benchmark para o refinamento de hiperparâmetros que faremos mais adiante.

Analisando os plots de feature importance, também mostrados abaixo, vemos que a renda (Q006) é, como esperávamos, uma feature importantíssima para a previsão da nota de Matemática. Entretanto, dependendo da métrica que escolhemos para avaliar as importâncias relativas entre as features (weight, cover ou gain para o XGBoost e split ou gain para o LightBoost), obtemos resultados diferentes para as outras métricas. Por exemplo, usando o critério de weight, o XGBoost aponta que a escolaridade dos pais é mais importante que a própria renda. Mas, usando o critério do cover, o mesmo modelo aponta que a escolaridade é uma feature de pouca importância relativa. A significância do resultado dos valores de feature importances obtidos por meio dessas métricas é bastante contestada. Uma solução, apresentada [nesse artigo](https://towardsdatascience.com/interpretable-machine-learning-with-xgboost-9ec80d148d27), é a utilização da métrica SHAP (sigla para Shapley Additive Explanations).

Com o SHAP, encontramos um resultado coerente entre os dois modelos. A renda é a feature mais importante. No segundo pelotão, temos sexo, escolaridade dos pais e renda per capita. Com cerca de metade da importância desse último, temos a quantidade de carros (Q010) e a feature que informa se a raça é branca. Por fim, com menor  importância, temos quantidade de residentes (Q005), empregado doméstico (Q007), TV por assinatura (Q021) e as outras variáveis de raça.

# Feature Selection
Municiados com a EDA e as feature importances, vamos selecionar as features que permanecerão em nosso modelo final. TV por assinatura (Q021) e empregado doméstico (Q007) apresentaram importâncias relativamente pequenas, não as utilizaremos. O número de moradores na residência (Q005) não tem grande importância diretamente, mas participa do cálculo da renda per capita, que é variável importante. Portanto, permanece no modelo. Renda, sexo, escolaridade máxima entre os pais têm grande importância e também permanecerão.

Resta decidirmos sobre a raça. A classe dummy de raça mais importante, com pelo menos o dobro das outras raças, é a "branca". De fato, o que mais importa para o modelo é determinar se o candidato é branco ou não é. As classes de raça "amarela" e "indígena" são as menos importantes de todas as features. Uma ideia seria fazermos um feature engineering, criando a variável isBranco, que agregaria todas as raças não brancas numa única categoria. Mas isso implicaria, em produção, perguntar para novos entrevistados "qual é sua raça?", aumentando o grau de complexidade do questionário. Rodamos os mesmos modelos acima excluindo as features de raça, e vimos que o RMSE sobe cerca de 0.4%. Desse modo, nossa decisão é excluir as features de raça do modelo. Vamos arcar com esse prejuízo na sua acurácia em favor de um modelo mais simples.

Portanto, nosso modelo passa a ser descrito por renda, número de residentes, renda per capita, escolaridade máxima dos pais, sexo e número de carros.

Rodando um novo benchmark com o LightGBM, utilizando somente as features acima, encontramos um MAE de 75.0 e um RMSE de 92.9.

Comparando com o resultado obtido para o benchmark que incorporava todas as features que acabamos de eliminar (74.6 e 92.4, respectivamente), vemos que a simplificação do modelo nos custou uma piora de ~0.5% no RMSE.

Ficamos bastante satisfeitos com esse trade-off entre simplifidade e acurácia. Na verdade, ficamos tão satisfeitos que nos perguntamos qual seria o efeito sobre o RMSE se removêssemos a feature "quantidade de carros", de menor importância entre as features que sobreviveram. Nesse caso, nossos MAE e RMSE passam para 75.2 e 93.2, respectivamente. Tivemos uma perda relativa de ~0.86% em relação ao benchmark inicial, que incorporava todas as features de antes do feature selection.

Assim, nosso modelo final será descrito apenas por renda, número de residentes, renda per capita, escolaridade máxima dos pais e sexo.

```python
X_final = X3_labels[['Q005', 'Q006', 'TP_SEXO', 'max_escol', 'perCapita', 'Q010']].copy()
X_final.head()
y = model_data2['NU_NOTA_MT']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.3, random_state=101)

import lightgbm as lgb

params = {}
#params['learning_rate'] = 0.01
#params['objective'] = 'regression'
params['metric'] = 'rmse'
params['verbose'] = 0

#incluindo 'carros'
train_data = lgb.Dataset(X_train, label=y_train)
validation_data = lgb.Dataset(X_test, label=y_test)
bst = lgb.train(params, train_data, 1000, valid_sets=validation_data, early_stopping_rounds=5, verbose_eval=False)

#avaliando métricas do modelo com carros
predictions = bst.predict(X_test)
print('incluindo a feature quantidade de carros:')
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))

#excluindo 'carros'
X_final = X3_labels[['Q005', 'Q006', 'TP_SEXO', 'max_escol', 'perCapita']].copy()
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.3, random_state=101)

train_data = lgb.Dataset(X_train, label=y_train)
validation_data = lgb.Dataset(X_test, label=y_test)
bst = lgb.train(params, train_data, 1000, valid_sets=validation_data, early_stopping_rounds=5, verbose_eval=False)

#avaliando métricas do modelo com carros
predictions = bst.predict(X_test)
print('\nexcluindo a feature quantidade de carros:')
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))

sns.distplot((y_test-predictions),bins=50)
```

# Refinando o Modelo
Escolhemos o LightGBM para nosso modelo final porque teve acurácia bastante parecida com o XGBoost, mas é mais rápido de ser treinado. Além disso, um ponto negativo do XGBoost é que a biblioteca python onde está implementado é mais pesada, o que geraria [problemas](https://medium.com/@AlexeyButyrev/xgboost-on-aws-lambda-345f1394c2b) quando fôssemos fazer o deploy serverless do modelo no AWS Lambda.

A implementação do Gradient Boosting do LightGBM possui vários parâmetros de treinamento, como por exemplo o learning_rate (coeficiente do resultado encontrado por cada árvore para o cálculo da previsão final) e max_depth (número máximo de camadas de cada árvore, que restringe a quantidade de splits que cada árvore faz). Esses parâmetros não são treináveis, i.e., devem ser informados manualmente ao algoritmo antes que ele realize o treinamento que irá determinar as especificações do modelo que fará a previsão de novos exemplos. Por isso, eles são chamados de hiperparâmetros.

O processo de refinamento consiste em treinar diferentes modelos, que diferem pelos seus hiperparâmetros, e utilizar nosso conjunto de validação para verificar qual dos modelos faz a melhor previsão. Compararemos a qualidade dos modelos pela métrica do erro quadrático médio ("RMSE"). Nosso benchmark mostrou RMSE de 93.2 na previsão da nota.

As duas formas mais comuns de se testar diferentes conjuntos de hiperparâmetros são o "grid search" e o "random search". No primeiro, definimos um conjunto de valores discretos candidatos para cada hiperparâmetro, e treinamos modelos com todas as combinações possíveis entre os diferentes hiperparâmetros. No segundo, definimos intervalos de valores candidatos para cada hiperparâmetro, e selecionamos aleatoriamente, dentro de cada intervalo, os valores que serão utilizados em cada modelo.

Uma terceira forma de fazer a busca por hiperparâmetros é o [Bayesian optimization](https://towardsdatascience.com/automated-machine-learning-hyperparameter-tuning-in-python-dfda59b72f8a). Nesse método, também sorteamento aleatoriamente valores para os hiperparâmetros. Mas, ao contrário do random search, onde a distribuição de probabilidades entre os valores possíveis é uniforme, o Bayesian optimization utiliza iterativamente os resultados que vão sendo obtidos para explorar mais intensamente os intervalos de valores mais promissores. Assim ,a cada novo modelo treinado, é atualizada a distribuição de probabilidade dos valores associados a cada hiperparâmetro. A implementação do Bayesian optimization que utilizaremos aqui é a do pacote [hyperopt](https://github.com/hyperopt/hyperopt).

Decidimos fazer um split de train/dev/test de 80/10/10%. Isso coloca mais de 360 mil exemplos (cada) no dev e test sets, valor suficiente para que cumpram sua função (avaliar o melhor entre dois modelos e estimar a performance do algoritmo em exemplos nunca vistos, respectivamente). Abaixo, verificamos a distribuição de notas para o train, dev e test set, a fim de verificarmos se são semelhantes. Do contrário, poderíamos estar treinando um algoritmo que faria overfitting numa certa região de notas.

O código abaixo implementa a busca pelos hiperparâmetros ótimos usando o Bayesian optimization em 500 modelos.

```python
##exportando os datasets finais
#X_final = X3_labels[['Q005', 'Q006', 'TP_SEXO', 'max_escol', 'perCapita']].copy()
#y_final = model_data2['NU_NOTA_MT']

#y_final.to_csv('y_final.csv', index=False)
#X_final.to_csv('X_final.csv', index=False)
```

```python
#Função Objetivo
import lightgbm as lgb
from hyperopt import STATUS_OK

X_final = X3_labels[['Q005', 'Q006', 'TP_SEXO', 'max_escol', 'perCapita']].copy()
y_final = model_data2['NU_NOTA_MT']

#split 80/10/10
X_train, X_val, y_train, y_val = train_test_split(X_final, y_final, test_size=0.1, random_state=101)
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.1111111111, random_state=101)


train_set = lgb.Dataset(X_train, y_train)
val_set = lgb.Dataset(X_val, y_val)
test_set = lgb.Dataset(X_test, y_test)

def objective(params):
    """Objective function for Gradient Boosting Machine Hyperparameter Tuning"""
    
    params['metric'] = 'rmse'
    
    params['subsample'] = params['boosting_type']['subsample']
    boost_type = params['boosting_type']['boosting_type']
    del params['boosting_type']
    params['boosting_type'] = boost_type
    
    params['num_leaves'] = int(params['num_leaves'])
    params['min_child_samples'] = int(params['min_child_samples'])
    params['subsample_for_bin'] = int(params['subsample_for_bin'])
    
    
    
    
    
    eval_metrics = {}
    bst = lgb.train(params, train_set, 1000, valid_sets=val_set, early_stopping_rounds=10, verbose_eval=False, evals_result=eval_metrics, valid_names=['validation_data'])
  
    # Extract the best score
    loss = eval_metrics['validation_data']['rmse'][len(eval_metrics['validation_data']['rmse']) - 1]
    
    
    
    # Dictionary with information for evaluation
    return {'loss': loss, 'params': params, 'status': STATUS_OK}
```

```python
f, ax = plt.subplots()

sns.distplot(y_train, label='train')
sns.distplot(y_val, label='dev')
sns.distplot(y_test, label='test')
plt.legend()
plt.title('train/dev/test set possuem mesma distribuição de notas')
plt.show()
f.savefig("trantestdev2", bbox_inches='tight', dpi=500)
```

```python
#Espaço de hiperparâmetros a ser explorado
from hyperopt import hp

# Define the search space
space = {
    'boosting_type': hp.choice('boosting_type', [{'boosting_type': 'gbdt', 'subsample': hp.uniform('gdbt_subsample', 0.5, 1)}, 
                                                 {'boosting_type': 'goss', 'subsample': 1.0}]),
    'num_leaves': hp.quniform('num_leaves', 20, 150, 1),
    'learning_rate': hp.loguniform('learning_rate', np.log(0.01), np.log(0.3)),
    'subsample_for_bin': hp.quniform('subsample_for_bin', 20000, 300000, 20000),
    'min_child_samples': hp.quniform('min_child_samples', 20, 500, 5),
    'reg_alpha': hp.uniform('reg_alpha', 0.0, 1.0),
    'reg_lambda': hp.uniform('reg_lambda', 0.0, 1.0),
    'colsample_bytree': hp.uniform('colsample_by_tree', 0.6, 1.0)
}
```

```python
#teste: sorteando um conjunto de hiperparâmetros
from hyperopt.pyll.stochastic import sample

# Sample from the full space
example = sample(space)

# Dictionary get method with default
subsample = example['boosting_type'].get('subsample', 1.0)

# Assign top-level keys
example['boosting_type'] = example['boosting_type']['boosting_type']
example['subsample'] = subsample

example
```

```python
#Fazendo a busca
MAX_EVALS = 500

from hyperopt import tpe
# Algorithm
tpe_algorithm = tpe.suggest

from hyperopt import Trials
# Trials object to track progress
bayes_trials = Trials()

from hyperopt import fmin

# Optimize
best = fmin(fn = objective, space = space, algo = tpe.suggest, 
            max_evals = MAX_EVALS, trials = bayes_trials)
```

# Interpretando e Explicando o Modelo

Para entender o motivo por trás das previsões que nosso modelo faz para cada candidato, vamos usar os valores SHAP, calculados usando a biblioteca shap para python.

Primeiramente, vamos entender melhor o que são esses valores de SHAP. Para cada feature de cada candidato, é calculado um valor SHAP. A tabela abaixo mostra os valores SHAP para todas as cinco features  dos cinco primeiros candidatos. Além disso, a sexta coluna dá a soma dos valores de SHAP para aquele candidato (a soma das 5 primeiras colunas). A sétima coluna mostra qual é a previsão de nota que nosso modelo fez para aquele candidato.

```python
import shap

#porção do train_set usada para estudar o SHAP
X_train_for_shap = X_train

#é necessário declararmos explicitamente o parâmetro 'objective' para a biblioteca SHAP saber que tarefa estamos realizando
modelo_final.params['objective'] = 'regression'

#calculando os valores de SHAP para cada feature de cada candidato
explainer = shap.TreeExplainer(modelo_final)
shap_values = explainer.shap_values(X_train_for_shap)

#valores de shap para cada usuário
shap_df = pd.DataFrame(shap_values, columns=X_train.columns)

#somando os shap values de todas as features de cada candidato
shap_sum = np.round(shap_df.sum(axis=1),2)

#fazendo a previsão da nota para cada candidato
predictions = np.round(modelo_final.predict(X_train_for_shap),2)

#criando dataframe com os shaps, shap_sum e prediction para cada ponto amostral
df = pd.concat([np.round(shap_df,2), pd.Series(shap_sum, name='SHAP_sum'), pd.Series(predictions, name='prediction')], axis=1)
df.head()
```

Para entender a relação dos valores SHAP com a previsão que nosso modelo fez para certo candidato, plotamos abaixo a soma dos SHAP de certo candidato contra a previsão de nota que nosso modelo fez para aquele candidato.

Perceba que existe uma relação (linear) monotonamente crescente entre essas grandezas. Isso significa que um valor de SHAP positivo faz o modelo aumentar a previsão de nota, e um valor de SHAP negativo faz o modelo diminuir a previsão da nota. O SHAP de uma certa feature, para um certo candidato, informa a contribuição marginal que aquela feature tem na previsão de nota!

Por exemplo, o primeiro candidato (índice 0) da tabela acima tem um valor negativo, de -26.49, para a coluna Q006 (renda). Isso quer dizer que, o fato desse candidato pertencer àquela classe de renda, fez com que nosso modelo diminuísse a previsão de nota (e a queda na previsão é proporcional ao módulo desse valor de SHAP 26.49).
Já o segundo candidato (índice 1) da tabela acima tem valor de SHAP para a renda de +37,97. Isso significa que nosso modelo aumentou a previsão de nota para aquele candidato devido a ele pertencer àquela classe de renda.
